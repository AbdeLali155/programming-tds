```python
import numpy as np
import numpy.random as rd
import matplotlib.pyplot as plt
```

```python
def AfficherImg(img):
    img=plt.imread(img)
    plt.axis("off")
    plt.imshow(img,cmap='gray')
```

```python
AfficherImg("lenna.png")
```

```python
def ImageNoire(L,C):
    M=np.zeros((L,C))
    plt.axis("off")
    plt.imshow(M,cmap='gray')
```

```python
ImageNoire(5,5)
```

```python
def Imageblanche(h,l):
    M=np.ones((h, l))
    plt.axis("off")
    plt.imshow(M,cmap='gray',vmin=0, vmax=1)
```

```python
Imageblanche(5,5)
```

```python
def imageBlancheNoire(L,C):
    img=np.zeros((L, C))
    for i in range(L):
        for j in range(C):
            img[i][j]=(i+j)%2
    plt.axis("off")
    plt.imshow(img,cmap='gray',vmin=0, vmax=1)
```

```python
imageBlancheNoire(6,6)
```

```python
def image_gris(L, C):
   """Construit la matrice L*C d'une image en mode gris."""
   img=np.zeros((L, C))
   for i in range(L):
       for j in range(C):
           img[i][j]=rd.randint(0,256)
   plt.imshow(img,cmap='gray')
```

```python
image_gris(2, 4)
```

```python
def DrapeauFr(H,L):
    noire = [0, 0, 0]
    jaune= [255, 255, 0]
    rouge = [255, 0 , 0]
    France = [noire]*H + [jaune]*H + [rouge]*H
    France = [France]*L
    Fr= np.array(France)
    plt.axis("off")
    plt.imshow(Fr)
    
```

```python
DrapeauFr(2, 2)
```
